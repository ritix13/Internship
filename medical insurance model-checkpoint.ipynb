```python
a.replace({'sex':{'male':1,'female':2}},inplace=True)
```

```python
a.replace({'smoker':{'yes':1,'no':2}},inplace=True)
```

```python
a
```

```python
x=a.drop(columns='charges',axis=1)
y=a['charges']
```

```python
plt.figure(figsize=(20,15),facecolor='yellow')
plotnumber=1
for column in x:
    if plotnumber<=6:
        ax=plt.subplot(2,3,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column)
        plt.ylabel('charges')
    plotnumber+=1
plt.show()
```

```python
scaler=StandardScaler()     # when the data in dataset has different scale for that we have to maek the same scale , 
                             #for which we can use standardscalers
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
 x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
regression=LinearRegression()
regression.fit(x_train,y_train)    #model is trained
```

```python
regression.score(x_train,y_train)
```

```python
regression.score(x_test,y_test)
```
